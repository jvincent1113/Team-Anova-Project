```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

salary = pd.read_csv("C:/Users/Vinny/Desktop/Vinny\School/Team-Anova-Project/Vinny_WIP/salary_data_states.csv")
```

```python
salary.groupby('JOB_TITLE')['PAID_WAGE_PER_YEAR'].mean().sort_values(ascending=False).head(10)

#bar plot
#overall market share
```

```python
salary['JOB_TITLE'].value_counts()
```

```python
salary['JOB_TITLE_SUBGROUP'].value_counts()
```

```python
salary.groupby('JOB_TITLE_SUBGROUP')['PAID_WAGE_PER_YEAR'].sum()
```

```python
colors = ['deepskyblue','lightsalmon', 'tomato', 'yellow', 'plum', 'aqua', 'springgreen', 'violet']

salary.groupby('JOB_TITLE_SUBGROUP')['PAID_WAGE_PER_YEAR'].sum().plot(kind='pie', autopct='%1.1f%%', startangle=80, colors=colors)
plt.title('Total Income by Job Type')
plt.ylabel('')
plt.show()
```

```python
colors = ['springgreen','tomato', 'deepskyblue', 'violet', 'yellow', 'lightsalmon', 'plum', 'aqua']

salary['JOB_TITLE_SUBGROUP'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=200, colors=colors)
plt.title('Types of Jobs Held by Immigrants')
plt.ylabel('')
plt.show()
```
