```python
import csv

# Function to read CSV data
def read_csv(file_path):
    data = []
    with open(file_path, 'r', newline='', encoding='utf-8') as file:
        reader = csv.reader(file)
        for row in reader:
            data.append(row)
    return data

# File path for the "salary_data_states-csv" dataset
salary_data_states_path = r'C:\Users\miles\OneDrive\Desktop\Pyt_SData_Copy\salary_data_states-csv.csv'

# Read CSV data
salary_data_states_data = read_csv(salary_data_states_path)

# File path for the "COL_TAXES_STATES-csv" dataset
col_taxes_states_path = r'C:\Users\miles\OneDrive\Desktop\Pyt_SData_Copy\COL_TAXES_STATES-csv.csv'

# Read CSV data
col_taxes_states_data = read_csv(col_taxes_states_path)
```
