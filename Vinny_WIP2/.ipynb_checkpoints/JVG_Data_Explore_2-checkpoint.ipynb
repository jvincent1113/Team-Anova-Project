```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

salary = pd.read_csv("C:/Users/Vinny/Desktop/Vinny/School/Team-Anova-Project/Vinny_WIP/salary_data_states.csv")

salary.head()
```

```python
# What states (of those I am willing to move to) have the highest paying data-related salaries? (VINNY)

datasal = salary.loc[salary['JOB_TITLE_SUBGROUP']=='data analyst']

datasal.groupby('WORK_STATE')['PAID_WAGE_PER_YEAR'].mean().round(2).sort_values(ascending=False)
```

```python
# Differences between job sub-categories?

datasal.groupby('JOB_TITLE_SUBGROUP')['PAID_WAGE_PER_YEAR'].mean().round(2)

```

```python
# Which companies have the highest salaries for those sub-types?

# Data Scientist
sciensal = salary.loc[salary['JOB_TITLE_SUBGROUP']=='data scientist']

sciensal.groupby('EMPLOYER_NAME')['PAID_WAGE_PER_YEAR'].mean().head(5).round(2).sort_values(ascending=False)

```
